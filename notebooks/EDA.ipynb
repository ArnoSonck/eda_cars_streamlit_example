# Vehicles' EDA

## Libraries and data

```python
import pandas as pd
import plotly.graph_objects as go
import streamlit as st

df = pd.read_csv('../vehicles_us.csv')
```

## Exploration

```python
df.info()
```

### Year exploration

```python
# Clean model_year column
years = df["model_year"].dropna()

# Ensure numeric
years = pd.to_numeric(years, errors="coerce").dropna()

# Basic stats
mean_year = years.mean()
median_year = years.median()

print(f"Mean year: {mean_year:.2f}")
print(f"Median year: {median_year:.0f}")
```

```python
# Histogram
fig = go.Figure(data=[go.Histogram(x=df["model_year"])])
fig.update_layout(title_text='Year Distribution')
# fig.show()
```

```python
filtered_df = df[~df['model_year'].isna()]

fig = go.Figure(data=[go.Scatter(x=filtered_df['model_year'], y=filtered_df['price'], mode='markers')])

fig.update_layout(title_text='Relation year - price')

fig.show()
```
